```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import  seaborn as sns
%matplotlib inline
```

```python
athlete= pd.read_csv('athlete_events.csv')
regions =pd.read_csv('noc_regions.csv')
```

```python
goldMedals = athletes_df[( athletes_df.Medal=='Gold')]
goldMedals.head()
```

```python
goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]
```

```python
goldMedals['ID'][goldMedals['Age']>60].count()
```

```python
sporting_event = goldMedals['Sport'][goldMedals['Age']>60]
sporting_event
```

```python
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('Gold medals over 60 years')
```

```python
goldMedals.region.value_counts().reset_index(name='Medal').head()
```

```python
# Getting the top 5 countries with the most gold medals
totalGoldMedals = goldMedals.region.value_counts().reset_index(name='Medal').head(6)

# Creating a bar plot
g = sns.catplot(x="region", y="Medal", data=totalGoldMedals,
                height=5, kind="bar", palette="rocket")

g.despine(left=True)  # Remove left spine for aesthetics

g.set_xlabels("Top 5 Countries")
g.set_ylabels("Number of Medals")

plt.title("Gold Medals per Country")
plt.show()
```
